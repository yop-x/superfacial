## 1.1 Loop through face pics


```python
import os
import pandas as pd
import numpy as np
```

```python
yop_yes = [
    'Aaron Donald',
    'Alex Pall',
    'Chadwick Boseman',
    'Chris Hemsworth',
    'Dan Raynolds',
    'Iamdapower',
    'James Corden',
    'Leo Chang',
    'Matt Severson',
    'Miguel Silvestre',
    'P Elephant',
    'Taylor Rapp',
    'Tylehner',
    ''
]
```

```python
import os 
path = "/Users/yopxue/Desktop/projects/super_facial/50_cele/faces/"

faces = {}


for person in os.listdir(path):
    person_dir = os.path.join(path, person)

    if not os.path.isdir(person_dir):
        continue

    faces[person] = []
    for imagename in os.listdir(person_dir):
        image_path = os.path.join(person_dir, imagename)

        image = face_recognition.load_image_file(image_path)
        encodings = face_recognition.face_encodings(image)

        if len(encodings) > 0:
            faces[person].append(encodings[0])

        else:
            print(f'no face found in {image_path}.')



    
```

```python
for i in faces:
    print(i, len(faces[i]))
```

```python
rows = []

for person, encs in faces.items():
    for enc in encs:
        row = {'name': person}

        for i, value in enumerate(enc):
            row[f"f{i}"] = value

        rows.append(row)
df = pd.DataFrame(rows)
```

```python
df
```

```python
# assign yes to yop_yes 

df['target'] = df['name'].isin(yop_yes).astype(bool)
```

```python
# normalization 

from sklearn.preprocessing import normalize
```

```python
X = df.filter(regex="^f").values
X = normalize(X, norm="l2")
```

```python
np.linalg.norm(X, axis=1).min(), np.linalg.norm(X, axis=1).max()
```

```python
# Binary target
y = df["target"].astype(int).to_numpy()   # 1=True, 0=False

# Group by identity to avoid leakage
groups = df["name"].to_numpy()
```

```python
from sklearn.model_selection import GroupShuffleSplit

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups=groups))

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(
    max_iter=2000,
    class_weight="balanced",  # helpful if yop_yes is smaller
    solver="lbfgs"
)

clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

pred = clf.predict(X_test)
proba = clf.predict_proba(X_test)[:, 1]

print("Accuracy:", accuracy_score(y_test, pred))
print("ROC AUC:", roc_auc_score(y_test, proba))
print(classification_report(y_test, pred, digits=4))
```
